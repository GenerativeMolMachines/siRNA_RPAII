## Generation of RDKit descriptors

```python
import json

sense_column, antisense_column = "Sense", "AntiSense"

def get_max_sequence_length(dataframe: pd.DataFrame, sense_column: str = sense_column, antisense_column: str = antisense_column):

    max_legth = 0
    for sense_seq, antisene_seq in zip(dataframe[sense_column].to_list(), dataframe[antisense_column].to_list()):
        sense_seq, antisene_seq = json.loads(sense_seq), json.loads(antisene_seq)
        max_length_for_pair = max(len(sense_seq), len(antisene_seq))
        if max_legth < max_length_for_pair:
            max_legth = max_length_for_pair
    return max_legth

max_sequence_length = get_max_sequence_length(df)
print(f"Max sequence length: {max_sequence_length} nucleotides")
```

### Getting NEW RDKit descriptors

```python
class RDKit_new:

    @classmethod
    def encode_sequences(cls, sequences_list, max_length=max_sequence_length) -> dict:

        descriptor_names = list(rdMolDescriptors.Properties.GetAvailableProperties())
        get_descriptors = rdMolDescriptors.Properties(descriptor_names)
        num_descriptors = len(descriptor_names)

        x = np.empty((0, num_descriptors*max_length), float)

        skipped_ids = []
        for i, seq in enumerate(sequences_list):
            descriptors_set = np.empty((0, num_descriptors), float)
            skipped = False
            for smiles in seq:
                if skipped is False:
                    molecule = Chem.MolFromSmiles(smiles)
                    try:
                        descriptors = np.array(get_descriptors.ComputeProperties(molecule))
                    except:
                        print("\033[93m", f"Unable to generate descriptors for SMILES: {smiles}", "\033[0m")
                        skipped = True
                if skipped is False:
                    descriptors = descriptors.reshape((-1, num_descriptors))
                else:
                    descriptors = np.empty((0, num_descriptors), float)
                descriptors_set = np.append(descriptors_set, descriptors, axis=0)
            
            if skipped is True:
                skipped_ids.append(i)
                print("\033[91m", f"Skipped id: {i}", "\033[0m")

            scaled_df = pd.DataFrame(descriptors_set, columns=descriptor_names)
            while scaled_df.shape[0] < max_sequence_length:
                scaled_df.loc[scaled_df.shape[0]] = [0] * num_descriptors

            array = np.array(scaled_df)
            array = array.flatten()
            x = np.vstack((x, array))
        print(f"shape: {x.shape}")
        print(f"skipped ids: {skipped_ids}")

        return {"x": x, "skipped_ids": skipped_ids}

    @classmethod
    def get_dataset(cls, df: pd.DataFrame, max_length=max_sequence_length):

        senses = [json.loads(line) for line in df[sense_column]]
        antisenses = [json.loads(line) for line in df[antisense_column]]

        max_sequence_length = 27

        print("Generating descriptors for senses")
        sense_response = cls.encode_sequences(sequences_list=senses, max_length=max_sequence_length)
        print("Generating descriptors for antisenses")
        antisense_response = cls.encode_sequences(sequences_list=antisenses, max_length=max_sequence_length)

        skipped_ids = sense_response.get("skipped_ids", [])  + antisense_response.get("skipped_ids", [])

        x_senses = sense_response.get("x", [])
        x_antisenses = antisense_response.get("x", [])

        smdb_ids = df["SMDBid"].to_list()
        print(f"len senses: {len(x_senses)}, len antisenses {len(antisenses)}, len SMDBids {len(smdb_ids)}")

        x = list()
        skipped_smdb_ids = []
        for i, smdbd_id in enumerate(smdb_ids):
            if i in skipped_ids:
                skipped_smdb_ids.append(smdbd_id)
                continue
            m = np.hstack((x_senses[i], x_antisenses[i]))
            x.append(m)

        x = np.vstack(x)

        pd.DataFrame(x).to_csv("data/new_rdkit_descriptors.csv", index=False, header=False)


        conc = df["siRNA concentration"].to_list()
        efficacy = df["Efficacy, %"].to_list()
        trans_duration = df["Duration after transfection"].to_list()
        experiment = df["Experiment used to check activity"].to_list()
        target_gene = df["Target gene"].to_list()
        cell = df["Cell or Organism used"].to_list()
        trans_method = df["Transfection method"].to_list()

        experiment_data = np.empty((0, 6))
        for i in range(len(smdb_ids)):
            if i in skipped_ids:
                continue
            line = np.array([
                conc[i], trans_duration[i], experiment[i],
                target_gene[i], cell[i], trans_method[i]
            ])

            experiment_data = np.vstack((experiment_data, line))
        
        pd.DataFrame(experiment_data).to_csv("data/new_rdkit_experiment_data.csv", index=False, header=False)

        y = np.empty((0, 1))
        for i in range(len(smdb_ids)):
            if i in skipped_ids:
                continue
            y = np.vstack((y, efficacy[i]))

        pd.DataFrame(y).to_csv("data/new_rdkit_targets.csv", index=False, header=False)

        print("\n"+"descriptors.shape:", x.shape)
        print("experiment_data.shape:", experiment_data.shape)
        print("target.shape:", y.shape, end='\n\n')

        print("skipped_cases:", skipped_smdb_ids)

        assert x.shape[0] == experiment_data.shape[0] == y.shape[0]

        return x, experiment_data, y
    
sequences_descriptors, experiment_data, target = RDKit_new.get_dataset(df)
```
